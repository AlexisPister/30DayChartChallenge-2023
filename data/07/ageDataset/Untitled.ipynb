```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
import vegafusion as vf
vf.enable()
```

```python
path = "AgeDataset-V1.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
set(df["Manner of death"])
```

```python
df_filt = df[df.Occupation == "Artist"]
df_filt = df_filt[df_filt["Birth year"] > 1900]
# df_filt = df_filt[df_filt["Birth year"] > 1940]
df_filt = df_filt[df_filt["Age of death"] < 31]
# df_filt = df_filt[df_filt["Manner of death"] != "Canserbero"]
```

```python
df_filt = df_filt.dropna(subset=['Manner of death'])
```

```python
df_filt["Age of death"] = df_filt["Age of death"].astype(int)
```

```python
substrings = ["suicide", "accident", "natural c"]
df_filt = df_filt[df_filt['Manner of death'].str.contains('|'.join(substrings))]
```

```python
df_filt['Manner of death'] = df_filt['Manner of death'].replace({"accident; natural causes": "accident", "suicide; accident": "suicide", "suicide; homicide": "suicide", "suicide; unfortunate accident": "suicide"})
```

```python
alt.Chart(df_filt).mark_bar(
    # color=alt.Color('Manner of death:N', scale=alt.Scale(domain=['accident', 'suicide'], range=['#1f77b4', '#ff7f0e']))
).encode(
    # x=alt.X("Age of death:Q", bin=alt.Bin(step=1, extent=[18, 30])),
    x=alt.X("Age of death:O"),
    y='count()',
    # color='Manner of death',
    color=alt.Color('Manner of death:N', scale=alt.Scale(domain=['accident', "natural causes", 'suicide'], range=["#67a9cf","#016c59", "#1c9099"]), legend=alt.Legend(title='Manner of death')),
    tooltip='Manner of death'
).properties(
    # title='Distribution of the age of death of artist celebrities born after 1900',
    title='artist celebrities born after 1900 age of death distribution from accident, natural causes, and suicide causes.',
    width=800,
    height=600
).configure_axis(
    labelFontSize=14,
    titleFontSize=16,
    # titleAngle=0,
    labelAngle=0
).configure_title(
    fontSize=22
)
```
